# **Intro to Data Type**

In this section, we will explore one of the core concepts of Python programming: **"Data types."** Data types refer to the kinds of data used in programming, and Python offers a variety of data types. **These data types act as the fundamental building blocks of a program, and the way data is handled differs depending on the type.**

This section introduces the basic data types provided in Python, including **Numeric, Boolean, Sequence, Dictionary,and Set**. We will explain the characteristics and usage of each data type. ## **1. Numeric Type**

In Python, numeric types can be used in various forms such as
- **1) integers**,
- **2) floats**,
- **3) complex**.

These types are widely used in programming to represent and manipulate numerical data using mathematical operations.


### **1.1 Integer**
- Integers are used to represent **whole numbers**, **including negative numbers, zero, and positive numbers.** Examples include: 42, -7, 0

```python
#integer

num1 = 42
num2 = -7
print(num1 + num2)
```

### **1.2 Floats**
- Floats are used to represent **numbers with a decimal point**. Examples include: 3.14, -0.01, 1.23

```python
#float
num1 = 3.14
num2 = -0.01
print(num1 + num2)
```

### **1.3 Complex**
- The complex data type in python consists of two values, the first one is the **real part** of the complex number, and the second one is the **imaginary part** of the complex number.

```python
#complex

z1 = 2 + 3j
z2 = 1 + 4j

print(z1 + z2)
```

### **1.4 Arithmetic operations**

You can use **various arithmetic operations** related to numeric types
- Addition
- Subtraction
- Multiplication
- Division
- Floor Division
- Modulus
- Exponential

**1)** **Addition (+):**

```python
result = 5 + 3
print(result)
```

**2) Subtraction (-):**

```python
result = 10 - 4
print(result)
```

**3) Multiplication** (*)

```python
result = 7 * 2
print(result)
```

**4) Division (/):**

Divides the first number by the second, returning a floating-point result.

```python
result = 8 / 2
print(result)
```

**5) Floor Division (//)**

Divides the first number by the second, returning the integer part of the result (rounded down).

```python
result = 9 // 4
print(result)
```

**6) Modulus (%)**

Returns the remainder of the division between two numbers.

```python
result = 10 % 3
print(result)
```

**7) Exponentiation** (**)

Raises the first number to the power of the second number.

```python
result = 2 ** 3
print(result)
```

### **3.1 String**

- In Python, a string is a data type used to **represent a sequence of characters**.
- Strings are enclosed in either **single quotes (')** or **double quotes (")** and are versatile, **supporting operations such as concatenation, repetition, and slicing.**

#### **1) Representation**

- Strings can be created using single quotes (') or double quotes ("). Both are functionally identical.

```python
single_quote = 'Hello, Class!'

print(single_quote)
```

```python
double_quote = "Hello, Class!"

print(double_quote)
```

#### **2) Multiline Strings**

- For strings spanning multiple lines, use **triple quotes**

```python
multiline_string1 = '''Hello, Class!
My name is Phil.
I look forward to learning
Big Data and Machine Learning!'''

print(multiline_string1)
```

```python
multiline_string2 = """Hello, Class!
My name is Phil.
I look forward to learning
Big Data and Machine Learning!"""

print(multiline_string2)
```

#### **3) String Operations**

- Strings support various operations like **concatenation, repetition, and slicing**.

**Concatenation (+):** Combines two strings

```python
greeting = "Hello, " + "Class!"
print(greeting)
```

**Repetition**(*):  Repeats a string a specified number of times.

```python
repeated = "Hi! " * 3
print(repeated)
```

**Slicing ([]):** Extracts parts of a string.

```python
text = "Python"
print(text[0:3])
```

#### **4) Understanding String Length, Indexing, and Slicing in Python**

**Calculating String Length**

- The **len()** function in Python calculates the **length of a string, including spaces and special characters**.

```python
a = "I put my heart and my soul into my work, and have lost my mind in the process - Vincent Van Gogh"
print(len(a))
```

#### **5)  String Indexing**

- Indexing allows you to **access specific characters** in a string using their positions.
- Python uses **zero-based indexing**, **meaning the first character has an index of 0.**

```python
a = "Great things are done by a series of small things brought together - Vincent Van Gogh"
print(a[0])
print(a[13])
print(a[-1])
```

#### **6) String Slicing**

- Slicing **extracts a portion of a string** using the syntax string[start:end].
- The slice starts from the start index and includes characters **up to, but not including**, the end index.
- If you **omit start:** The slice **begins at the start of the string**.

```python
a = "Life isn't about finding your self. Life is about creating yourself -George Bernard Shaw"
print(a[0:35])
```

```python
print(a[:8])
print(a[11:])
print(a[:])
```

#### **7) Using Negative Indices in Slicing**

- Negative indices can also be used in slicing to access portions of **a string from the end.**

```python
print(a[11:-1])
```

#### **8) String Formatting**

- Starting from Python 3.6, **f-strings (formatted string literals)** provide a concise and readable way to embed expressions inside string literals.
- This feature allows you to include variables, computations, and other expressions directly within a string by prefixing it with f or F.

```python
keyword = "life"
name = "George Bernard Shaw"
print(f"{keyword} isn't about finding yourself. {keyword} is about creating yourself - {name}")
```

- The **format()** method in Python also provides a flexible way to format strings.

```python
print("{keyword} isn't about finding yourself. {keyword} is about creating yourself - {name}".format(keyword = "Life", name = "George Bernard Shaw"))
```

#### **9) String Built-in Functions in Python**

- Strings in Python come with a variety of built-in functions that allow you to manipulate and analyze them easily.
- These functions are called string methods, and you use them **by appending a dot (.) followed by the method name to a string variable.**



##### **a. Case Conversion**

**lower():** Converts all characters to **lowercase**.

```python
text = "Hello Class"
print(text.lower())
```

**upper():** Converts all characters to **uppercase**.

```python
print(text.upper())
```

**capitalize():** **Capitalizes the first character** of the string.

```python
print(text.capitalize())
```

**title():** Capitalizes the **first character of each word.**

```python
print(text.title())
```

**swapcase():** **Swaps the case of each character**.

```python
print(text.swapcase())
```

##### **b. Whitespace Management**

**strip():** The strip function **removes all consecutive spaces** (or other specified characters) **from both sides of a string**.

```python
text = "   Hello Class   "
print(text.strip())
```

**lstrip():** The lstrip function **removes all consecutive spaces** (or other specified characters) **from the left side **of a string. The "l" in lstrip stands for "left".

```python
print(text.lstrip())
```

**rstrip:** The rstrip function **removes all consecutive spaces** (or other specified characters) **from the right side** of a string. The "r" in rstrip stands for "right".

```python
print(text.rstrip())
```

##### **c. String Search and Replacement**

**find(substring):** **Returns the first index** of the substring, or -1 if not found.

```python
text = "Hello, Class!"
print(text.find("Class"))
```

**replace(old, new):** Replaces occurrences of old with new.

```python
print(text.replace("Class", "Phil"))
```

**count(substring):** **Counts the number of occurrences** of a substring.

```python
print(text.count("l"))
```

**split(delimiter):** **Splits** the string into a list of substrings **based on a delimiter.**

```python
text = "apple,banana,cherry"
print(text.split(","))
```

- Lists are **WIDELY USED** for tasks involving data processing, analysis, and organization, making them one of the most versatile and commonly used data types in Python.

#### **3) Nested Lists**

- Lists can **contain other lists as elements.**

#### **7) Modifying a List**

**Adding elements:**
- The **append()** method is used to add an element to the end of a list.

```python
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)
```

**Removing elements:**
- The remove() method is used to remove an element from a list.

```python
numbers.remove(2)
print(numbers)
```

**Changing elements:**

```python
numbers[0] = 10
print(numbers)
```

**insert(index, item):** **Inserts** an element **at a specific position.**


- The first argument (1 in this case) is the index where you want to insert the new element. Remember, Python list indices start at 0, so 1 refers to the second position in the list.
- The second argument ("blueberry") is the element you want to insert.

```python
fruits = ["apple", "banana", "cherry"]
fruits.insert(1, "blueberry")
print(fruits)
```

**index(item):** Returns the index of the first occurrence of the specified item.

```python
fruits = ["apple", "banana", "cherry"]
index = fruits.index("banana")
print(index)
```

**count(item):** Returns the **number of times the specified item appears** in the list.

```python
numbers = [1, 2, 3, 2, 4, 2]
count = numbers.count(2)
print(count)
```

**sort():** **Sorts the list in ascending order.**

```python
#ascending
numbers = [4, 2, 3, 1]
numbers.sort()
print(numbers)
```

```python
#decending
numbers = [4, 2, 3, 1]
numbers.sort(reverse=True)
print(numbers)
```

**reverse():** **Reverses the order** of the list.

```python
numbers = [1, 2, 3, 4]
numbers.reverse()
print(numbers)
```

**clear():** Removes all items from the list.

```python
fruits = ["apple", "banana", "cherry"]
fruits.clear()
print(fruits)
```

### **3.3 Tuple Type**

A **tuple** **is an immutable, ordered collection of elements** in Python, often used to group related data. **Once created, the elements of a tuple cannot be modified, added, or removed.**

Tuples are defined **using parentheses ()** or can be created without any brackets, separated by commas.

Tuples and lists are advantageous in different situations depending on the use case.

**Situations where using a list is beneficial:**
- When the values of elements are likely to change.
- When elements need to be added or removed.
- When the order of elements is important (e.g., sorting is required).
- When list-specific methods (e.g., append, remove, insert, etc.) need to be utilized.

**Situations where using a tuple is beneficial:**
- When the values of elements must remain unchanged (immutability is required).

**Choosing the appropriate data type between lists and tuples is essential based on the context.**


#### **1) Tuple**

## **4. Dictionary Type**

In Python, a **dictionary** is a mutable, unordered data structure composed of **key-value pairs**.

Dictionaries are created using curly braces **{}**or the **dict()** function. They provide various methods that make data manipulation easy and versatile.

**Key Features:**
- **Structure:** Composed of **keys (unique)** and **corresponding values**.
- **Mutable:** You **can add, modify, or delete key-value pairs**.

### **4.1 Create a Dictionary**



```python
# Using curly braces
my_dict1 = {'apple': 1, 'banana': 2, 'orange': 3}
print(my_dict1)
```

```python
# Using the dict() Function
my_dict2 = dict(apple=1, banana=2, orange=3)
print(my_dict2)
```

### **4.2 Adding Key-Value Pairs**

- To add a key-value pair, use the format:

```python
# Initial dictionary
dic = {'name': 'Tom', 'age': 25}
print(dic)

# Adding a new key-value pair
dic['city'] = 'New York'
print(dic)
```

### **4.3 Deleting Key-Value Pairs**

- To delete a specific key-value pair, use the del keyword:

```python
# Initial dictionary
dic = {'name': 'Tom', 'age': 25, 'city': 'New York'}
print(dic)

# Deleting a key-value pair
del dic['age']
print(dic)
```

### **4.4 Accessing Values Using Keys**

- To get a value from a dictionary, use the key inside square brackets ([]).

```python
# Dictionary of student grades
student_grades = {"Alice": 95, "Bob": 85, "Cindy": 90}

# Access Alice's grade
alice_grade = student_grades["Alice"]
print(alice_grade)
```

### **4.5 ) Dictionary Functions**

- Dictionaries come with several built-in functions that make it easy to process and explore key-value pairs.
- Commonly used functions include keys(), values(), items(), get(), clear(), and update()

**a. Getting Keys: keys()**
- The keys() function returns all the keys in a dictionary as a dict_keys object. This can be converted into a list if needed.

```python
d = {'apple': 3, 'banana': 2, 'orange': 4}
d_keys = d.keys()
print(d_keys)
print(list(d_keys))
```

**b. Getting Values: values()**

- The values() function **returns all the values** in a dictionary as a dict_values object. Like keys, these can be converted into a list.

```python
d_values = d.values()
print(d_values)
print(list(d_values))
```

**c. Getting Key-Value Pairs: items()**

- The items() function returns all key-value pairs in the dictionary as a dict_items object.
- Each pair is a tuple (key, value). These pairs can also be converted into a list or used directly in loops.

```python
fruit_stock = {"apple": 10, "banana": 5, "orange": 8}
all_items = fruit_stock.items()
print(all_items)
print(list(all_items))
```

**d. Using Key to Get Value (get)**

- The get() function in a dictionary retrieves the value associated with a given key.
- If the key does not exist, it can return a default value instead of raising an error, making it safer than direct access using [].

```python
beethoven = {
    "full_name": "Ludwig van Beethoven",
    "birth_year": 1770,
    "death_year": 1827,
    "nationality": "German"
    }

# Accessing data
print(beethoven["full_name"])
```

**e. Updating a Dictionary (update)**

- The update() function allows you to merge one dictionary into another.
- It adds all key-value pairs from one dictionary to the other, updating the values of overlapping keys.

```python
b = {
    "genres": "Classical",
    "notable_works": ["Symphony No. 9 (Ode to Joy)", "Moonlight Sonata", "Violin Concerto in D Major"]
    }

beethoven.update(b)
print(beethoven)
print(beethoven["notable_works"])
```

**f. Clearing a Dictionary (clear)**

- The clear() function removes all key-value pairs from a dictionary, leaving it empty.

```python
# Clear the dictionary
beethoven.clear()
print(beethoven)
```

## **5. Set Type**

The set data type in Python is **a collection of unique elements** that are **unordered and do not allow duplicates**.


There are two ways to create a set:

- The first method is to list elements separated by commas inside curly braces **{}**.
- The second method is to use the **set()** function to convert a sequence data type, such as a list, tuple, or string, into a set.

```python
set1 = {1, 2, 2, 2, 4, 5}
print(set1)
type(set1)
```

```python
# Converting a list to a set
my_list = [1, 2, 2, 2, 4, 5]
my_set = set(my_list)
print(my_set)
type(my_set)
```

```python
# Converting a string to a set
my_string = "hellooooooooooooo"
my_set = set(my_string)
print(my_set)
```

**Key Takeaways:**
- Sets are collections of unique, unordered elements.
- They are ideal for removing duplicates or performing mathematical operations like union, intersection, and difference.
- Sets cannot contain mutable objects like lists, but they can contain immutable objects like numbers, strings, or tuples.
